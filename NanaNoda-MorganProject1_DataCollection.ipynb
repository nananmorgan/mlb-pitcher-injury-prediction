## Data Collection

After spending a few hours trying to figure out the download from the Statcast Search page on the Baseball Savant website, I realized that I needed a different solution for collecting the data because the website was limiting the rows I could download, and it would have taken me hundreds of separate downloads to get the data I need. So instead, I am using an API to pull pitch-level data. And even with this new method, since a baseball season often has over 700,000 pitches, I will be reducing the seasons to 2021-2025 and I'll save each season into its own CSV file.  Also on the choice of seasons, our injury data starts in 2022, but we need a 30-day window prior to that to look back on, so including September 2021 will give us that coverage.

  
For the Injury data, I have been doing more digging and decided that I will be using an MLB Stats API as well, which replaces the web scraping approach. This is actually an improved approach from a fagile web scraping set up, and will be more reproducible in the future.

```python
#Import necessary libraries
import pandas as pd
import requests ##to make API calls
import time ##use to add time between API calls
import os as os #to create folders/check for files
from pybaseball import statcast #connects to Baseball Savant and download pitch data
```

```python
#Create folder 
os.makedirs("data/raw", exist_ok=True)

#Create the list of seasons we want to download
season_name_list = ["2021_buffer", "2022", "2023", "2024", "2025"]
start_date_list =  ["2021-09-01", "2022-03-31", "2023-03-30", "2024-03-20", "2025-03-20"]
end_date_list =    ["2021-11-05", "2022-11-06", "2023-11-02", "2024-11-03", "2025-11-03"]

#Show what I'm going to download
print("Seasons to download:")
print("-" * 50)
for i in range(len(season_name_list)):
    print(f"  {season_name_list[i]}: {start_date_list[i]} to {end_date_list[i]}")
```

```python
# Now this is the actual download step. It should check if the file already is in the folder 
# I created above before starting (just incase I need to rerun this at any time).

for i in range(len(season_name_list)):
    season = season_name_list[i]
    start = start_date_list[i]
    end = end_date_list[i]
 
    #Build the file path where we'll save this season's data
    file_path = "data/raw/statcast_" + season + ".csv"
 
    #Check if we already downloaded this season
    if os.path.exists(file_path):
        print(f"✓ {season} already downloaded, skipping.")
        continue
 
    #Download the data
    print(f"Downloading {season} ({start} to {end})...")
    print("  This may take 10-30 minutes. Please be patient.")
 
    try:
        # This is the main download call
        pitch_data = statcast(start_dt=start, end_dt=end)
 
        # Save to CSV file
        pitch_data.to_csv(file_path, index=False)
 
        # Pring progress
        row_count = len(pitch_data)
        print(f"  ✓ Done! Saved {row_count:,} rows to {file_path}")
 
    except Exception as error:
        print(f"  ✗ Error downloading {season}: {error}")
        print(f"  You can re-run this cell to try again.")
 
print("\nAll Statcast downloads complete")
```

```python
#Check if download was successful
print("Statcast Download Summary:")
print("-" * 60)
 
total_rows = 0
 
for i in range(len(season_name_list)):
    season = season_name_list[i]
    file_path = "data/raw/statcast_" + season + ".csv"
 
    if os.path.exists(file_path):
        # Read the file to count rows and pitchers
        df = pd.read_csv(file_path, low_memory=False)
        row_count = len(df)
        pitcher_count = df["pitcher"].nunique()
        total_rows = total_rows + row_count
        print(f"  {season}: {row_count:,} rows, {pitcher_count} unique pitchers")
    else:
        print(f"  {season}: FILE NOT FOUND — need to re-download")
 
print(f"\nTotal rows across all seasons: {total_rows:,}")
 
```

Now we'll move onto the Injury List data, which will serve as our target variable in predicting when pitchers will end up on the IL. This data is form the MLB public Stats API, and it returns the data in JSON format. 

```python
#I'll create a function to call this API so we can reuse it for each year.

def get_il_transactions(start_date, end_date):
    """
    Calls the MLB Stats API to get Injured List transactions.
    
    Parameters:
        start_date and end_date in "MM/DD/YYYY" format (example: "01/01/2022")

    Returns:
        A list of transaction records (each record is a dictionary)
    """
    #The URL for the MLB transactions API
    url = "https://statsapi.mlb.com/api/v1/transactions"
    
    #the parameters I'm sending with our request
    #this tells the API what kind of transactions I want
    params = {
        "startDate": start_date,
        "endDate": end_date,
        "transactionTypes": "Injured List,Reinstatement"
    }
    
    #make the API call
    response = requests.get(url, params=params)
    
    #Check if the call was successful
    if response.status_code != 200:
        print(f"  API Error: status code {response.status_code}")
        return []
    
    #convert the JSON into a Python dictionary
    data = response.json()
    
    #pull out the list of transactions
    # .get() returns an empty list [] if "transactions" key doesn't exist
    transactions = data.get("transactions", [])
    
    return transactions
 
print("Function 'get_il_transactions' is ready")
```

```python
#Now I'll be creating another function to clean up the data a bit. 
#The data is nested (dictionaries within dictionaries), so this will take all the fields and turn it into a flat dictionary

def clean_one_transaction(transaction):
    """
    Take one raw transaction from the API and pull out the fields we need.
    
    Parameters:
        transaction: a dictionary from the API response
    
    Returns:
        A simple, flat dictionary with just the fields we care about
    """
    #get the player data
    #The player data is nested inside a "player" key
    #if that key doesn't exist, we get an empty dictionary {}
    player_info = transaction.get("player", {})
    
    #Player's position is nested like player -> primaryPosition -> abbreviation
    position_info = player_info.get("primaryPosition", {})
    
    #get team info
    #IL placements have a "toTeam" (the team the player is going to the IL for)
    #IL activations might have "fromTeam" instead
    to_team = transaction.get("toTeam", {})
    from_team = transaction.get("fromTeam", {})
    
    #Use whichever team info is available
    if len(to_team) > 0:
        team_info = to_team
    else:
        team_info = from_team
    
    #get transaction type description
    #This is also nested as typeDesc -> description
    type_desc_info = transaction.get("typeDesc", {})
    
    #build a clean record
    clean_record = {
        "transaction_id": transaction.get("id"),
        "date": transaction.get("date"),
        "effective_date": transaction.get("effectiveDate"),
        "type_desc": type_desc_info.get("description", ""),
        "description": transaction.get("description", ""),
        "player_id": player_info.get("id"),
        "player_name": player_info.get("fullName", ""),
        "position": position_info.get("abbreviation", ""),
        "team_id": team_info.get("id"),
        "team_name": team_info.get("name", ""),
    }
    
    return clean_record
 
print("Function 'clean_one_transaction' is ready")
```

```python
years = [2022, 2023, 2024, 2025] #What I want to download

# This list will hold all the cleaned records across all years
all_records = []
 
for year in years:
    #Build the date range strings for this year
    start = "01/01/" + str(year)
    end = "12/31/" + str(year)
    
    print(f"Fetching IL transactions for {year}...", end=" ")
    
    #Call the function to get the raw transactions
    raw_transactions = get_il_transactions(start, end)
    
    #Clean each transaction and add to our list
    for txn in raw_transactions:
        clean_record = clean_one_transaction(txn)
        all_records.append(clean_record)
    
    print(f"✓ Found {len(raw_transactions)} transactions")
    
    #Add a one second buffer before the next API call
    time.sleep(1)
 
print(f"\nTotal transactions collected: {len(all_records)}")
```

```python
data_folder = "data/raw"
 
all_pitcher_ids = []
 
statcast_files = [
    "statcast_2021_buffer.csv",
    "statcast_2022.csv",
    "statcast_2023.csv",
    "statcast_2024.csv",
    "statcast_2025.csv",
]
 
for file_name in statcast_files:
    file_path = os.path.join(data_folder, file_name)
    if os.path.exists(file_path):
        df = pd.read_csv(file_path, usecols=["pitcher"])
        ids = df["pitcher"].unique().tolist()
        all_pitcher_ids.extend(ids)
        print(f"  {file_name}: {len(ids)} pitchers")
 
# Remove dupes
unique_pitcher_ids = list(set(all_pitcher_ids))
print(f"\nTotal unique pitcher IDs: {len(unique_pitcher_ids)}")
```

```python
def get_player_transactions(player_id):
    """
    Get all transactions for one player from the MLB Stats API.
    Returns a list of transaction dictionaries, or empty list on failure.
    """
    url = "https://statsapi.mlb.com/api/v1/transactions"
    params = {"playerId": player_id}
    
    try:
        response = requests.get(url, params=params, timeout=10)
        if response.status_code != 200:
            return []
        data = response.json()
        return data.get("transactions", [])
    except Exception as error:
        return []
 
print("Download function ready")
```

```python
all_il_records = []
total_pitchers = len(unique_pitcher_ids)
pitchers_with_il = 0
 
for i in range(total_pitchers):
    player_id = unique_pitcher_ids[i]
    
    #Get all transactions for player
    transactions = get_player_transactions(player_id)
    
    #Look through transactions for IL-related ones
    for txn in transactions:
        description = txn.get("description", "")
        description_lower = description.lower()
        
        #Check if this is an IL transaction
        is_il = (
            "injured list" in description_lower
            or "disabled list" in description_lower
        )
        
        if is_il:
            #Get player info
            person_info = txn.get("person", {})
            
            #Get team info
            to_team = txn.get("toTeam", {})
            from_team = txn.get("fromTeam", {})
            if len(to_team) > 0:
                team_info = to_team
            else:
                team_info = from_team
            
            #Build our record
            record = {
                "transaction_id": txn.get("id"),
                "date": txn.get("date"),
                "effective_date": txn.get("effectiveDate"),
                "description": description,
                "player_id": person_info.get("id"),
                "player_name": person_info.get("fullName", ""),
                "team_id": team_info.get("id"),
                "team_name": team_info.get("name", ""),
            }
            all_il_records.append(record)
    
    #Track pitchers with IL history
    has_il = False
    for txn in transactions:
        desc = txn.get("description", "").lower()
        if "injured list" in desc or "disabled list" in desc:
            has_il = True
            break
    if has_il:
        pitchers_with_il = pitchers_with_il + 1
    
    #Print progress every 100 pitchers
    if (i + 1) % 100 == 0:
        print(f"  Processed {i + 1}/{total_pitchers} pitchers "
              f"({pitchers_with_il} with IL history, "
              f"{len(all_il_records)} IL records so far)")
    
    #Give it a small buffer between
    time.sleep(0.5)
 
print(f"\nDone! Processed {total_pitchers} pitchers.")
print(f"Pitchers with IL history: {pitchers_with_il}")
print(f"Total IL records: {len(all_il_records)}")
```

```python
## Turn into DF and make sure fields are filled in

il_df = pd.DataFrame(all_il_records)
 
# Convert dates
il_df["date"] = pd.to_datetime(il_df["date"])
il_df["effective_date"] = pd.to_datetime(il_df["effective_date"])
il_df["year"] = il_df["date"].dt.year
 
#Check that player info is not blank
print("Checking player_id and player_name columns:")
print(f"  player_id missing: {il_df['player_id'].isna().sum()} out of {len(il_df)}")
print(f"  player_name missing: {(il_df['player_name'] == '').sum()} out of {len(il_df)}")
 
print(f"\nSample of player info:")
print(il_df[["player_id", "player_name", "date", "description"]].head(10).to_string())
```

```python
#Create new fields by using the description text

#Extract action (placed vs activated)
action_list = []
for desc in il_df["description"]:
    desc_lower = desc.lower()
    if "placed" in desc_lower:
        action_list.append("placed")
    elif "activated" in desc_lower:
        action_list.append("activated")
    elif "transferred" in desc_lower:
        action_list.append("transferred")
    else:
        action_list.append("other")
 
il_df["action"] = action_list
 
#Get IL duration (10-day, 15-day, 60-day
il_type_list = []
for desc in il_df["description"]:
    desc_lower = desc.lower()
    if "60-day" in desc_lower:
        il_type_list.append("60-day")
    elif "15-day" in desc_lower:
        il_type_list.append("15-day")
    elif "10-day" in desc_lower:
        il_type_list.append("10-day")
    elif "7-day" in desc_lower:
        il_type_list.append("7-day")
    else:
        il_type_list.append("unknown")
 
il_df["il_type"] = il_type_list
 
#Get injury description
#The injury is usually the text after the last period in the description
#Example: "...on the 10-day injured list. Right elbow sprain."
#Try to capture: "Right elbow sprain"
injury_list = []
for desc in il_df["description"]:
    # Split on ". " to find the injury part
    parts = desc.split(".")
    
    # Look for the injury text (usually the last non-empty part)
    injury_text = ""
    for part in parts:
        part = part.strip()
        #Skip parts that contain "injured list" or "disabled list"
        if "injured list" in part.lower() or "disabled list" in part.lower():
            continue
        #Skip empty parts
        if len(part) == 0:
            continue
        #Skip parts that are just about the team/player (contain "placed" or "activated")
        if "placed" in part.lower() or "activated" in part.lower():
            continue
        #skip parts about retroactive dates
        if "retroactive" in part.lower():
            continue
        #Hopefully this is the injury description
        injury_text = part
    
    injury_list.append(injury_text)
 
il_df["injury"] = injury_list
 
print("New columns added: action, il_type, injury")
print()
```
